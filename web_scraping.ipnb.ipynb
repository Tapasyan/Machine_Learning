# Web Scraping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"
```

```python
r = requests.get(url)

print(r.text)
```

# Tags

```python
soup = BeautifulSoup(r.text,"lxml")
```

```python
prices = soup.find("h4",{"class":"price float-end card-title pull-right"})
```

```python
prices = soup.find_all("h4",class_ = "price float-end card-title pull-right")
for i in prices:
    print(i.text)
```

```python
names = soup.find_all("a",class_ = "title")
product_names = []
for i in names:
    name = i.text
    product_names.append(name)

print(product_names)
```

```python
prices = soup.find_all("h4",class_ = "price float-end card-title pull-right")
price = []
for i in prices:
    prices = i.text
    price.append(prices)
print(price)
```

```python
desc = soup.find_all("p",class_ = "description card-text")
product_desc = []
for i in desc:
    description = i.text
    product_desc.append(description)
print(product_desc)
```

```python
reviews = soup.find_all("p",class_ = "review-count float-end")
product_review = []
for i in reviews:
    review = i.text
    product_review.append(review)
print(product_review)
```

```python
df = pd.DataFrame({"Product name":product_names,"Price":price,"description":product_desc,"reviews":product_review})
df
```

```python
df[["Product name","Price"]].min()
```

```python
# Remove the dollar sign from the 'Price' column
df['Price'] = df['Price'].str.replace('$', '', regex=False).astype(float)
```

```python
df.groupby("Product name")["Price"].max().plot.bar(color = "cyan",edgecolor = "black")
plt.show()
```
